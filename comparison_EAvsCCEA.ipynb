**Comparaci√≥n de los resultados obtenidos entre el EA y el CCEA**

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
len(range(0,250,8))
```

```python
plt.figure(figsize=(10,6))

ev_ea = np.load("obj_evolution_EA.npy")
ev_ccea = np.load("obj_evolution_CCEA.npy")

plt.plot(range(0,len(ev_ea)), ev_ea, label='EA')
plt.xlabel('Iteraciones') 
plt.ylabel('Energ√≠a') 

plt.plot(range(0,len(ev_ccea)*8,8), ev_ccea, label='CCEA')
plt.xlabel('Iteraciones') 
plt.ylabel('Energ√≠a') 

plt.title('EA vs CCEA')
plt.legend()
plt.show()
```
